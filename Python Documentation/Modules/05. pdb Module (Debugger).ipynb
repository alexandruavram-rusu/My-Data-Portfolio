# pdb Module

Most times, especially when writing complex code, you are going to have an error. Fiding the error and solving it is crucial for the project. While exception handling with `print()` statements can be useful, most of the time it's not enough.

`pdb` is a built-in Python debugger module, allowing the implementation of an interactive debugging environment. It supports setting (conditional) breakpoints and single stepping at the source line level, inspection of stack frames, source code listing, and evaluation of arbitrary Python code in the context of any stack frame. It also supports post-mortem debugging and can be called under program control.

[Documentation](https://docs.python.org/3/library/pdb.html)

```python
import pdb
```

Lets create an error on purpose, like trying to add a number with a list.

```python
a = 3
b = 2
c = [1,2,3,4,5]

result = a + b
print(result)

result2 = b + c
print(result2)
```

Let's implement a `set_trace()` using the pdb module. This will allow us to basically pause the code at the point of the trace and check if anything is wrong. 

Make sure it's placed before a line which might cause an error.

```python
a = 3
b = 2
c = [1,2,3,4,5]

result = a + b
print(result)

#Set trace before error line
pdb.set_trace()

result2 = b + c
print(result2)
```
