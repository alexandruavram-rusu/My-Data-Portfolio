## Chained Comparison Operators

Sometimes you might need to compare multiple outputs at the same time. This is called a chained comparison. There are two statements that help Python in this: *and* & *or*


### and

In order for the output to be True, both statements have to be True. If either one or both is False, the output is false.

```python
#When one is False 

(1 >= 2) and (2 <= 3)
```

```python
#When both are True

(1 <= 2) and (2 <= 3)
```

### or

In order for the output to be True, only one statement has to be True. The output is False only if both statements are False.

```python
(1 >= 2) or (2 <= 3)
```

```python
(1 >= 2) or (2 >= 3)
```
